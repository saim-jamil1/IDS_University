```python
# male female ratio

male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']

Total_males = male['Gender'].count()
Total_female = female['CGPA'].count()


# Pie chart to show ratio 

Gender = [Total_males,Total_female]
mylabels = ["Male", "Female"]
mycolors = ["blue", "Green"]

plt.pie(Gender, labels = mylabels, colors = mycolors)
plt.show() 
```

```python

# Plot the CGPA of all male students on a histogram with intervals 2.0-2.5, 2.6-3.0, 3.1-3.5, 3.6-4.0.
Total_males_cgpa = male['CGPA']
intervals = [2.0, 2.5, 3.0, 3.5, 4.0]
plt.hist(Total_males_cgpa, bins=intervals)
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.show() 
```

```python
# Plot the HSSC-1 marks of all male vs female students on a scatter plot

HSSC_1_males = male['HSSC-1']
HSSC_1_female = female['HSSC-1']
marks_range = random.sample(range(0, 500), 21)


plt.scatter(HSSC_1_males, marks_range, marker='o')
plt.scatter(HSSC_1_female, marks_range, marker='x')
plt.xlabel('Obtained Marks', fontsize=12)
plt.ylabel('Total Marks', fontsize=12)
plt.legend()
plt.show()
```
